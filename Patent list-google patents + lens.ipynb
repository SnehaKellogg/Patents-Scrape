```python
list0 = []
list1=[]
list2=[]
list3 =[]
company_list=[]
```

```python
for lensid in lensids:
    ids = lensid # --define this in the loop used or here for this to work. We will use it from the list of cvs we exported earlier. 
    url = baseurl + ids + ext
    browser.visit(url)
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    legalinf = soup.find_all('div', id= 'legal')
    legalinfo = legalinf[0]
    heads1 = legalinfo.find('div', class_='div-table-heading')#.get_text(strip = True)
    if heads1 == None:
        entry = {'lensid': ids, 'url': url, header[0]: 'NA', header[1]: "NA", header[2]: "NA", header[3]: 'NA', 'entrynumber': "NA"}
        company_list.append(entry)
        print('done with this ID')
        continue
    heads = heads1.find_all('div', class_='div-table-cell')
    n=0
    header = []
    for head in heads:
        line = head.get_text(strip = True)
        header.append(line)
        n+=1
    values = legalinfo.find_all('div', class_='div-table-cell-mobile')#.get_text(strip = True)
    value_number = 0
    #run it for all rows
    for value in values:
        line = value.get_text(strip = True)
        if value_number%4 == 0:
            list0.append(line)
            col0 = line
        elif value_number%4 == 1:
            list1.append(line)
            col1 = line
        elif value_number%4 == 2:
            list2.append(line)
            col2 = line
        else:
            list3.append(line)
            col3 = line
            entry = {'lensid': ids, 'url': url, header[0]: col0, header[1]: col1, header[2]: col2, header[3]: col3, 'entrynumber': value_number}
            company_list.append(entry)
            time.sleep(.5)
        value_number+=1
    time.sleep(3)
        
        
#  check if different sheets to be made,
# also try and except clause for all of this. need to check where it stops using one. 
```

```python
bausch_list = pd.DataFrame(company_list)
```
